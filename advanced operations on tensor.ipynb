```python
#Creating 1st dummy tesnor.
import torch
a = torch.rand(10,10)
print("a:", a)
```

```python
#Creating 2nd dummy tensor
b = torch.rand(10, 10)
print("b:", b)
```

```python
#Matrix multiplicatin of a and b --- a x b
result=torch.matmul(a , b )
print (result)
```

```python
#Converting 2d tensor to 1d an doing dot product --- a.b
A1=torch.flatten(a)
B1=torch.flatten(b)
Dot_product=torch.dot(A1 , B1 )
print("Dot_prodect is :", Dot_product)
```

```python
#Creating another dummy data set, reshaping dummy data in two different sizes
test = torch.tensor([4,4,2,5])
print("test:",test)
c=test.reshape(2,2)
print("c:" , c)
d=test.reshape(4,1)
print("d:", d)
```
